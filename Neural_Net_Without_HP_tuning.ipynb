# Data Preprocessing

Dropping columns with non-numerical values to ease plotting.

Additionally, changing Y and N values in 'pha' feature to 0 and 1 for numerical values.

Dropping all rows with NaN diameter values, along with converting non-NaN diameter values to float

```python
sel_data = data.drop(columns=["name", "spec_B", "spec_T", "neo", "G", "extent", "IR", "condition_code", "n_obs_used", "data_arc"])
sel_data['pha'] = sel_data['pha'].map({'Y': 1, 'N': 0})
sel_data['class']=sel_data['class'].map({'MBA':0, 'OMB':1, 'MCA':2, 'AMO':3, 'IMB':4, 'TJN':5, 'CEN':6, 'APO':7, 'ATE':8,'AST':9, 'TNO':10, 'IEO':11, 'HYA':12, 'PAA':13})
display(sel_data["class"].unique())
#Dropping rows with missing diameter values
sel_data = sel_data.dropna(subset=['diameter'])
sel_data.dtypes
print(sel_data["class"].unique())
```

```python
#Converting diameter values from object to float
sel_data['diameter'] = sel_data['diameter'].str.replace('km', '').astype(float)
sel_data
```

```python
# I want to get rid of the very big values in a and diameter that might mess up the normalization of the data
sel_data = sel_data[sel_data['a'] < 10]
sel_data = sel_data[sel_data['diameter'] < 200]
```

```python
#Normalizing Data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sel_data_norm = pd.DataFrame(scaler.fit_transform(sel_data), columns = sel_data.columns)

display(sel_data)
display(sel_data_norm)
```

## Linear Regression Model to predict ``moid`` using ``diameter``


```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(sel_data_norm, test_size=0.2, random_state=42)


from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from scipy import stats


X_train = train_df[['diameter']]
y_train = train_df['moid']


X_test = test_df[['diameter']]
y_test = test_df['moid']

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred_new = model.predict(X_train)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error: {mse}")
```

## Neural Network for predicting ``moid``

```python
sel_data_neural = sel_data_norm.drop(columns = ['a','om','w','q','ad','pha','per_y','rot_per','GM','BV','UB','class','n','per','ma'])
```

```python
sel_data_neural.columns
sel_data_neural = sel_data_neural.dropna(subset=['albedo'])

```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(sel_data_neural, test_size=0.2, random_state=42)

X_train = train_df.drop(columns=['moid'])
y_train = train_df['moid']

X_test = test_df.drop(columns=['moid'])
y_test = test_df['moid']
```

```python
classifier = Sequential()
classifier.add(Dense(units = 32, activation = 'relu', input_dim = X_train.shape[1]))
classifier.add(Dense(units = 16, activation = 'relu'))
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 4, activation = 'relu'))
classifier.add(Dense(units = 2, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'relu'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
classifier.summary()
```

```python
history = classifier.fit(train_df.astype(float).drop(columns = ['moid']), train_df.astype(float)['moid'], batch_size = 32, epochs = 10)
```

```python
yhat_test = classifier.predict(test_df.drop(columns = ['moid']))
```

```python
yhat_test
```

```python
y_test = test_df['moid'].astype(float)
y_test
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, yhat_test)
print("Neural-MSE",mse)

#Calculating training MSE
yhat_train = classifier.predict(train_df.drop(columns = ['moid']))
mse_train = mean_squared_error(train_df['moid'], yhat_train)
print("Training-MSE",mse_train)
```
