# Data Preprocessing

Dropping columns with non-numerical values to ease plotting.

Additionally, changing Y and N values in 'pha' feature to 0 and 1 for numerical values.

Dropping all rows with NaN diameter values, along with converting non-NaN diameter values to float

```python
sel_data = data.drop(columns=["name", "spec_B", "spec_T", "neo", "G", "class", "extent", "IR", "condition_code", "n_obs_used", "data_arc"])
sel_data['pha'] = sel_data['pha'].map({'Y': 1, 'N': 0})

#Dropping rows with missing diameter values
sel_data = sel_data.dropna(subset=['diameter'])
sel_data.dtypes
```

```python
#Converting diameter values from object to float
sel_data['diameter'] = sel_data['diameter'].str.replace('km', '').astype(float)
sel_data
```

```python
# I want to get rid of the very big values in a and diameter that might mess up the normalization of the data
sel_data = sel_data[sel_data['a'] < 10]
sel_data = sel_data[sel_data['diameter'] < 200]
```

```python
#Normalizing Data

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sel_data_norm = pd.DataFrame(scaler.fit_transform(sel_data), columns = sel_data.columns)

display(sel_data)
display(sel_data_norm)
```
